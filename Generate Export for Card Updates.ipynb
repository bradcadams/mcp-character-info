```sql
SELECT
   c.character_id
  ,CASE WHEN a.alter_ego_count > 1
    THEN c.character_name || ' (' || c.alter_ego || ')'
    ELSE c.character_name
    END character_name
  ,c.threat
  ,c.front_health stamina_front
  ,c.back_health stamina_back
  ,c.extra_power
FROM
  mcp.game.characters c
  LEFT JOIN 
    (
      SELECT
         c2.character_name
        ,count(distinct c2.alter_ego) alter_ego_count
      FROM
        mcp.game.characters c2
      WHERE
        c2.current_version
      GROUP BY
        ALL
    ) a
    ON
      c.character_name = a.character_name
WHERE
  c.current_version
```
