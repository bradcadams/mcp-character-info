```sql
SELECT 
   c.character_id
  ,c.character_name
  ,c.alter_ego
  ,c.current_version
  ,c.effective_date
  ,CASE 
    WHEN c.current_version THEN NULL 
    ELSE LAG(c.effective_date) 
      OVER(PARTITION BY c.character_id ORDER BY c.effective_date DESC) - 1
    END end_date
  ,c.front_health healthy_stamina
  ,c.back_health injured_stamina
  ,c.front_health + coalesce(c.back_health,0) total_stamina
  ,c.threat
FROM
  mcp.game.characters c
ORDER BY
   c.character_id
  ,c.effective_date
```

```sql
SELECT DISTINCT
   c.character_id
  ,ac.affiliation_name
FROM
  mcp.game.characters c
  LEFT JOIN mcp.game.affiliated_charcters ac
    ON c.character_id = ac.character_id
ORDER BY
   c.character_id
  ,ac.affiliation_name
```
